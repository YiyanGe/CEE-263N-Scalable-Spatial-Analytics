```python
import pandas as pd
import itertools
import numpy as np
from scipy import linalg

import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
import datetime

import math
from sklearn import mixture
from sklearn.externals.six.moves import xrange
```

## 1. Read and clean data

```python
df = pd.read_table('new_york\Mildred.csv',header=None)  # error_bad_lines=False
len(df)
```

```python
def clean_data(data):
    data = data.apply(lambda x: x.str.split(','),axis=1)
    data['time'] = data[0].apply(lambda x: x[0])
    data['lat'] = data[0].apply(lambda x: float(x[1]))
    data['long'] = data[0].apply(lambda x: float(x[2]))
    return data[['time','lat','long']]
```

```python
# filter weekend 
def select_weekday(data):
    data['weekday'] = data['time'].apply(lambda x: datetime.datetime.strptime(x, '%Y-%m-%d %H:%M:%S').weekday())
    d = data.loc[data.weekday.isin([0,1,2,3,4])] # fileter out twitts on weekends
    d['hour'] = d['time'].apply(lambda x: int(x[11:13]))
    d['min'] = d['time'].apply(lambda x: int(x[14:16]))
    d['time_of_day'] = d.hour + d['min'] / float(60)
    return d[['lat','long','time_of_day']]
```

```python
x = select_weekday(clean_data(df))
```

```python
fig = plt.figure(figsize= (16,6))
ax = plt.subplot(1, 2, 1)
plt.scatter(x['lat'], x['long'], s=50, c=x['time_of_day'] ,alpha=0.6)
plt.title('Location distribution')
plt.colorbar()

ax = plt.subplot(1, 2, 2)
plt.scatter(x['lat'], x['time_of_day'], s=50, alpha=0.6)
plt.show()
```

## 2. GMM Model

```python
## Fit a GMM
X = x.as_matrix()
gmm = mixture.GMM(n_components=5, covariance_type='full', n_iter=100)
title = "Expectation-maximization"
gmm.fit(X)

color_iter = itertools.cycle(['c', 'navy', 'cornflowerblue', 'gold','darkorange'])
Y_ = gmm.predict(X)  # array of labels, same length as x
```

```python
fig = plt.figure(figsize= (12,12))
for i, (mean, covar, color) in enumerate(zip(gmm.means_, gmm._get_covars(), color_iter)):
    v, w = linalg.eigh(covar)
    u = w[0] / linalg.norm(w[0])
    
    ax = plt.subplot(2, 1, 1)
    plt.scatter(X[Y_ == i, 0], X[Y_ == i, 1],50, color=color)
    plt.xlabel('Latitude')
    plt.ylabel('Longitude')
    
    ax = plt.subplot(2, 1, 2)
    plt.scatter(X[Y_ == i, 0], X[Y_ == i, 2],30, color=color)
    plt.xlabel('Latitude')
    plt.ylabel('Time of the day')
```

```python
x['label'] = Y_
x.label.value_counts()
```

```python
# compare time distribution of two components with largest weight
x['label'] = Y_
fig = plt.figure(figsize= (16,4))
# Time distribution of one component
plt.subplot(1, 2, 1)
plt.hist(x[x.label==1].time_of_day.values,bins=20,alpha=0.3)
plt.title('Time distribution')

# Time distribution of another component
plt.subplot(1, 2, 2)
plt.hist(x[x.label==2].time_of_day.values,bins=20,alpha=0.3)
plt.title('Time distribution')

plt.show()
```

## 3. Investigate whether a person is a commuter

Criteria to determine whether a person is a commuter :
1. weight of home/work shoud be much larger than other activities, each of them should be larger than 0.3.
2. time (hour of the day) of cluster identified as home/work  should follow a particular pattern (9am - 5pm work and 0 - 6am + 8pm - 11 pm home).
3. Standard deviation of location of cluster identified as home/work should be small 

```python
def get_summary(n,cov,m,w):
    sd_lat  =[];sd_long=[];sd_time=[]
    for i in range(0,n):
        # Each degree of latitude is approximately 69 miles (111 kilometers) apart
        a = math.sqrt(  np.diagonal(cov[i])[0] ) * 111 # in kilometers
        b = math.sqrt(  np.diagonal(cov[i])[1] ) * 111   # in kilometers
        c = math.sqrt(  np.diagonal(cov[i])[2] )  
        sd_lat.append(a)
        sd_long.append(b)
        sd_time.append(c)
        
    summary = pd.concat([pd.DataFrame(m,columns=['mean_lat','mean_long','mean_time']),
                         pd.DataFrame({"comp":range(0,n)}),
                         pd.DataFrame({"sd_lat":sd_lat}),
                         pd.DataFrame({"sd_long":sd_long}),
                         pd.DataFrame({"sd_time":sd_time}),
                         pd.DataFrame({"weight":w})
                         ],axis=1).sort(['weight', 'mean_time'], ascending=False)
    return(summary)
```

### Find coodinates of Home and Work/School + the most likely time a person is at Home and at Work/School

```python
import os

name = [];commute=[];home_lat=[];home_long=[];home_time=[]
work_lat=[];work_long=[];work_time=[]

for file in os.listdir("new_york"):
    if file.endswith(".csv"):
        name.append(file.split('.')[0])
        df = pd.read_table( 'new_york/'+ file ,header=None)  # error_bad_lines=False
        
        x = select_weekday(clean_data(df))
        X = x.as_matrix()
        gmm = mixture.GMM(n_components=4, covariance_type='full', n_iter=100)
        gmm.fit(X)
        Y_ = gmm.predict(X)  # array of labels, same length as x

        summary = get_summary(n= gmm.n_components, cov = gmm.covars_,m = gmm.means_,w = gmm.weights_)
        summary.to_csv('new_york/summary/'+ file.split('.')[0]+'_summary.csv',index=False)
        
        if sum(summary.weight > 0.3) >= 1 and sum(summary.weight > 0.2) >= 2:
            commute.append(1)
            home_lat.append(summary.loc[0].mean_lat);home_long.append(summary.loc[0].mean_long); 
            home_time.append(summary.loc[0].mean_time) 
            work_lat.append(summary.loc[1].mean_lat);work_long.append(summary.loc[1].mean_long); 
            work_time.append(summary.loc[1].mean_time)
        else:            
            commute.append(0)
            home_lat.append(0);home_long.append(0); home_time.append(0)
            work_lat.append(0);work_long.append(0); work_time.append(0)

summary_all = pd.concat([pd.DataFrame({"name":name}),pd.DataFrame({"commute":commute}),
                         pd.DataFrame({"home_lat":home_lat}),pd.DataFrame({"home_long":home_long}),
                         pd.DataFrame({"home_time":home_time}),pd.DataFrame({"work_lat":work_lat}),
                         pd.DataFrame({"work_long":work_long}),pd.DataFrame({"work_time":work_time}) ],axis=1)
summary_all.to_csv('new_york/' +'summary_all.csv',index=False)
```
